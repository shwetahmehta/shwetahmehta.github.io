```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import matplotlib.cm as cm
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
import pandas as pd
pd.set_option('display.width', 500)
pd.set_option('display.max_columns', 100)
pd.set_option('display.notebook_repr_html', True)
```

```python
"""
Read in train and test as Pandas DataFrames
"""
df_train = pd.read_csv("train.csv")
df_test = pd.read_csv("test.csv")
```

```python
df_train.head()
```

```python
df_test.head()
```

```python
#store gap values
Y_train = df_train.gap.values
#row where testing examples start
test_idx = df_train.shape[0]
#delete 'Id' column
df_test = df_test.drop(['Id'], axis=1)
#delete 'gap' column
df_train = df_train.drop(['gap'], axis=1)
```

```python
#DataFrame with all train and test examples so we can more easily apply feature engineering on
df_all = pd.concat((df_train, df_test), axis=0)
df_all.head()
```

```python
"""
Example Feature Engineering

this calculates the length of each smile string and adds a feature column with those lengths
Note: this is NOT a good feature and will result in a lower score!
"""
#smiles_len = np.vstack(df_all.smiles.astype(str).apply(lambda x: len(x)))
#df_all['smiles_len'] = pd.DataFrame(smiles_len)

#strList = ['c', 'nH', '=', '#', 'c1cs', 'c1c', 'SiH2', 'c45', 'ccc']
#counter = 1
#for val in strList: 
#    'new_f'+str(counter) = np.vstack(df_all.smiles.astype(str).apply(lambda x: x.count(val)))
#    counter+=1

new_f1 = np.vstack(df_all.smiles.astype(str).apply(lambda x: x.count('c')))
new_f2 = np.vstack(df_all.smiles.astype(str).apply(lambda x: x.count('nH')))
new_f3 = np.vstack(df_all.smiles.astype(str).apply(lambda x: x.count('=')))
new_f4 = np.vstack(df_all.smiles.astype(str).apply(lambda x: x.count('#')))
new_f5 = np.vstack(df_all.smiles.astype(str).apply(lambda x: x.count('c1sc')))
new_f6 = np.vstack(df_all.smiles.astype(str).apply(lambda x: x.count('c1c')))
new_f7 = np.vstack(df_all.smiles.astype(str).apply(lambda x: x.count('SiH2')))
new_f8 = np.vstack(df_all.smiles.astype(str).apply(lambda x: x.count('c45')))
new_f9 = np.vstack(df_all.smiles.astype(str).apply(lambda x: x.count('SiH2')))
new_f10 = np.vstack(df_all.smiles.astype(str).apply(lambda x: x.count('ccc')))
```

```python
df_all['new_f1'] = pd.DataFrame(new_f1)
df_all['new_f2'] = pd.DataFrame(new_f2)
df_all['new_f3'] = pd.DataFrame(new_f3)
df_all['new_f4'] = pd.DataFrame(new_f4)
df_all['new_f5'] = pd.DataFrame(new_f5)
df_all['new_f6'] = pd.DataFrame(new_f6)
df_all['new_f7'] = pd.DataFrame(new_f7)
df_all['new_f8'] = pd.DataFrame(new_f8)
df_all['new_f9'] = pd.DataFrame(new_f9)
df_all['new_f10'] = pd.DataFrame(new_f10)
```

```python
df_test_withSMFeatures = df_all[test_idx:]
df_test_withSMFeatures.to_csv('df_test_withSMFeatures.csv')
```
